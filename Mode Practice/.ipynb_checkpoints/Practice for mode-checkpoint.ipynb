```python
data1 = pd.read_csv('DOT.csv')
data1 = data1.fillna(np.nan)
```

```python
data1
```

```python
data1.sample(n=5)
```

```python
data1.sort_values(by='arr_delay', ascending=False)[:10]
```

```python
def is_delayed(x):
    return x > 0

data1['delayed'] = data1['arr_delay'].apply(is_delayed)
```

```python
data1['delayed'] = data1['arr_delay'].apply(lambda x: x > 0)
```

```python
data1['delayed'].value_counts()
```

```python
not_delayed = data1['delayed'].value_counts()[0] # first value of the result above
delayed = data1['delayed'].value_counts()[1] # second value of the result above
total_flights = not_delayed + delayed # total count of flights
print float(delayed) / total_flights # converting to float to get a float result
    
```

```python
data1['cancelled'].value_counts() # count the (0, 1) values
    
```

```python
not_delayed, delayed = data1['cancelled'].value_counts()
print delayed / (delayed + not_delayed), '<- without conversion'
print float(delayed) / (delayed + not_delayed), '<- _with_ conversion!'
    
```
