```python
df_order1 = df_order.rename(columns={'Net Base Item Amount USD': 'NetBaseItemAmount', 
                                     'Net Services Amount USD': 'NetServicesAmount',
                                    'Net Item Amount USD' : 'NetItemAmount',
                                    'Freight Amount USD' : 'FreightAmount',
                                    'Coupon Item Amount USD' : 'CouponItemAmount',
                                    'Coupon Amount USD' : 'CouponAmount',
                                    'Net Sales Amount USD' : 'NetSalesAmount'})
```

```python
#get rid of annoying dollar sign.
df_order1['NetBaseItemAmount'] = [x.strip('$') for x in df_order1.NetBaseItemAmount]
df_order1['NetServicesAmount'] = [x.strip('$') for x in df_order1.NetServicesAmount]
df_order1['NetItemAmount'] = [x.strip('$') for x in df_order1.NetItemAmount]
df_order1['FreightAmount'] = [x.strip('$') for x in df_order1.FreightAmount]
df_order1['CouponItemAmount'] = [x.strip('$') for x in df_order1.CouponItemAmount]
df_order1['CouponAmount'] = [x.strip('$') for x in df_order1.CouponAmount]
df_order1['NetSalesAmount'] = [x.strip('$') for x in df_order1.NetSalesAmount]
df_order1.head()
```

```python
#convert the order date into datetime and get the months.
df_order1 = df_order1.rename(columns={'Order Date': 'Order_Date'})
df_order1.Order_Date = pd.to_datetime(df_order1.Order_Date, format='%m/%d/%y')
df_order1['OrderMonth'] = df_order1['Order_Date'].map(lambda x: 100*x.year + x.month) 
df_order1['OrderYear'] = df_order1['Order_Date'].map(lambda x: x.year ) 
# get the months & years' of the order time
```

```python
#Now we can view the rentention and new customer composition.
month_grouped = df_order1.groupby(['OrderMonth','New/Repeat Flag (Brand)'])
month_grouped.size().unstack()
```

```python
#Calculation the retention rate.
month_groupeddf = pd.DataFrame(month_grouped.size().unstack())
month_groupeddf.reset_index()
month_groupeddf['Retention%'] = month_groupeddf.New/(month_groupeddf.Repeat + month_groupeddf.New) *100
month_retention = month_groupeddf[['Retention%']]
month_retention.head()
```

```python
month_groupeddf = month_groupeddf.drop(['Retention%'], axis = 1)
```

```python
month_groupeddf.plot(kind='barh', stacked=False, figsize=[15,25], colormap='spring')
```

```python
#Now we can view the rentention and new customer composition.
year_grouped = df_order1.groupby(['OrderYear','New/Repeat Flag (Brand)'])
year_grouped.size().unstack()
```

```python
#Calculation the retention rate.
year_groupeddf = pd.DataFrame(year_grouped.size().unstack())
year_groupeddf.reset_index()
year_groupeddf['Retention%'] = year_groupeddf.New/(year_groupeddf.Repeat + year_groupeddf.New) *100
year_retention = year_groupeddf[['Retention%']]
year_retention.head()
```
