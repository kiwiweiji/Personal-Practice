```python

#Let's split the date and time.


df['when_the_delivery_started'] = pd.to_datetime(df['when_the_delivery_started'])
df['when_the_Jumpman_arrived_at_pickup'] = pd.to_datetime(df['when_the_Jumpman_arrived_at_pickup'])
df['when_the_Jumpman_left_pickup'] = pd.to_datetime(df['when_the_Jumpman_left_pickup'])
df['when_the_Jumpman_arrived_at_dropoff'] = pd.to_datetime(df['when_the_Jumpman_arrived_at_dropoff'])

df['d_s_date'] = pd.to_datetime(df['when_the_delivery_started'].dt.date)
df['d_s_time'] = df['when_the_delivery_started'].dt.time
df['j_a_p_date'] = pd.to_datetime(df['when_the_Jumpman_arrived_at_pickup'].dt.date)
df['j_a_p_time'] = df['when_the_Jumpman_arrived_at_pickup'].dt.time
df['j_l_p_date'] = pd.to_datetime(df['when_the_Jumpman_left_pickup'].dt.date)
df['j_l_p_time'] = df['when_the_Jumpman_left_pickup'].dt.time
df['j_a_d_date'] = pd.to_datetime(df['when_the_Jumpman_arrived_at_dropoff'].dt.date)
df['j_a_d_time'] = df['when_the_Jumpman_arrived_at_dropoff'].dt.time

df.dtypes
```

```python
# A brief overview of our data.
df2.describe()
```

```python
# There are a lot of time nodes in the dataset. We choose the two with the least missing values
#to do some exploratory data analysis.
df2.columns
a_cols = ['delivery_id', 'customer_id', 'jumpman_id', 'vehicle_type',\
          'pickup_place', 'place_category', 'item_name', 'item_quantity',\
          'item_category_name', 'how_long_it_took_to_order', 'pickup_lat',\
          'pickup_lon', 'dropoff_lat', 'dropoff_lon', 'd_s_date', 'd_s_time',\
          'j_a_d_date','j_a_d_time']
df3 = df2[a_cols]
df3
```

```python
# Let do some feature engineering to prepare the future time analysis.

# Converting defferent obejects into time formats and preparing for the heatmap.

df2['when_the_delivery_started'] = pd.to_datetime(df2['when_the_delivery_started'],\
                                                  format="%Y-%m-%d %H:%M:%S.%f")
df2['when_the_Jumpman_arrived_at_dropoff'] = pd.to_datetime(df2['when_the_Jumpman_arrived_at_dropoff'],\
                                                  format="%Y-%m-%d %H:%M:%S.%f")

df3['d_s_wknum'] = df3['d_s_date'].dt.dayofweek
df3['d_s_wkd'] = df3['d_s_date'].dt.weekday_name
df3['d_s_mdnum'] = df3['d_s_date'].dt.day
df3['d_s_dh'] = df2['when_the_delivery_started'].dt.hour

df3['j_a_d_wknum'] = df3['j_a_d_date'].dt.dayofweek
df3['j_a_d_wkd'] = df3['j_a_d_date'].dt.weekday_name
df3['j_a_d_mdnum'] = df3['j_a_d_date'].dt.day
df3['j_a_d_dh'] = df2['when_the_Jumpman_arrived_at_dropoff'].dt.hour

df3.columns
```

```python
# We can visualize the number of deliveries grouping by different time frames.

df3_weekdays = df3.pivot_table(index=['d_s_wknum','d_s_wkd'],
                                  values='delivery_id',
                                  aggfunc='count')
df3_weekdays.plot(kind='bar', figsize=(8,6))
plt.ylabel('Total deliveries')
plt.title('deliveries by Week Day');

#Sundays have the highest amount of deliveries. It's weird that we might arbitrarily think the 
#delivery service might has higher needs on weekdays because people are busy and don't have time
#to cook or go to resturants.
```

```python
# We can explore if there is a latent pattern for each day in a month.

df3_monthdays = df3.pivot_table(index=['d_s_mdnum'],
                                  values='delivery_id',
                                  aggfunc='count')
df3_monthdays.plot(kind='bar', figsize=(8,6))
plt.ylabel('Total deliveries')
plt.title('deliveries by Month Day');

# It looks like for one month, it also follows the weekly pattern that we've discovered. 
```

```python
# Now it's the time for hours' distribution.

df3_hours = df3.pivot_table(index=['d_s_dh'],
                                  values='delivery_id',
                                  aggfunc='count')
df3_hours.plot(kind='bar', figsize=(8,6))
plt.ylabel('Total deliveries')
plt.title('deliveries by Hours');

# Obviously, during the lunch and dinner time, there are two peaks. But the dinner time has
#double amount of deliveries comparing to the lunch time. It could be there are more needs, or
#there are more jumpmen are available in that time frame.
```

```python
#Let's do some pivot research. We can group the data by Weekday and hour
summary = df3.groupby(['d_s_wkd', 'd_s_dh'])['delivery_id'].count()
#reset index
summary = summary.reset_index()
#convert to dataframe
summary = pd.DataFrame(summary)
#browse data
#rename last column
summary=summary.rename(columns = {'delivery_id':'Delivery_Counts', \
                                  'd_s_dh' : 'd_hours', 'd_s_wkd' : 'd_weekdays'}) 
summary.head()
```
